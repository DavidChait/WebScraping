# code using requests_html and no pyGoogleNews

## imports and constants

```python
from requests_html import HTMLSession
```

## the code

```python
url="http://news.google.com/rss/search?q=canadian%20federal%20election&hl=en-CA&gl=CA&ceid=CA:en"

session = HTMLSession()

response = session.get(url)

print(response.html.text)
```

```python
print(response.html.html)
```

```python
print(type(response.html))
print(type(response.content))
print(type(response.html.html))
print(dir(response.html.element()))
```

```python
links = response.html.element("link")
type(links[0])
```

```python
items = response.html.search_all("item")
for item in items:
    print(item.fixed)
    print(item.named)
    print(item.spans)
    # print(item, "\n")  <Result () {}> 
    
```

# code using requests and bs4

## constants and imports

```python
from bs4 import BeautifulSoup
import requests

URL="http://news.google.com/rss/search?q=canadian%20federal%20election&hl=en-CA&gl=CA&ceid=CA:en"
HEADER_DICT = {
               "Accept"  :  "*/*"  ,   
               "User-Agent"    : "Mozilla/5.0 (Windows NT 6.1; Win64; x64; rv:87.0) Gecko/20100101 Firefox/87.0" 
              }      
```

## the code

```python
response = requests.get(URL, headers=HEADER_DICT)
soup = BeautifulSoup(response.content, "html.parser")
#items = soup.find
#soup.contents
```

```python
titles = soup.find_all("title")
for title1 in titles:
    print(title1)
```
